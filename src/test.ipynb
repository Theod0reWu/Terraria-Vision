```python
import cv2
import matplotlib.pyplot as plt
```

```python
def display(img, path = None):
    '''
        Given a cv2 image, displays the image using pyplot
    '''
    if (len(img.shape) > 2 and img.shape[2] == 3):
        img = img[:, :, ::-1]
    else:
        plt.gray()
    plt.axis("off")
    plt.imshow(img)
    if (not path):
        plt.show()
    else:
        plt.savefig(path)
```

```python
def getSprite(img, x, y, incr = 16, offset = 2):
    start_x = x * 16 + offset * x
    start_y = y * 16 + offset * y
    return img[start_x: start_x + incr, start_y : start_y + 16]

def isBlank(img):
    return (img == 0).all()

def getDimensions(img, incr = 16, offset = 2):
    return img.shape[0] / (incr + offset), img.shape[1] / (incr + offset)
```

```python
path = "../images/terraria_tiles/Tiles_501.png"
```

```python
img = cv2.imread(path)
print(getDimensions(img))
display(img)
display(getSprite(img, 0,1))
print(isBlank(getSprite(img, 0, 1)))
```
